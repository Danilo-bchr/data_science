## Obtaining the Data


For this project, I will import the Pandas, Matplotlib and Seaborn libraries:

```
# importing the necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

The database for our project was obtained from the official Airbnb website. Let's import it via the link:


```
# importing the listings.csv file into a DataFrame
df = pd.read_csv("http://data.insideairbnb.com/canada/nb/new-brunswick/2022-06-24/visualisations/listings.csv")
```

### **Most expensive location in the dataset**



To find the most expensive location, I'm going to base myself on the *neighbourhood* variable and get the average price in each neighboruhood.

But first, I will make sure that the *price* variable doesn't have outliers.

I'll start by analyzing your main statistc data:

```
# analysis of the main statistical data
df_clean.price.describe()
```

It is possible to observe that the mean, 184.9, is closer to the 3rd quartile, 199, represented by 75% than the median, 143, represented by 50%.

Also, the maximum value (51000) is out of line considering that this is the price for one night's stay.

Therefore, it is possible to notice that we have the presence of outliers in the "*price*" variable. For example, the value of 51,000.

This can be confirmed in the box chart below:

```
# viewing the box plot
df_clean.price.plot(kind="box", vert=False, figsize=(15,3))
plt.show()
```

So that it is possible to clean the dataset, I will find out what the outliers are using the IQR Method:

I will start by calculating the value of the IQR, as well as the lower and upper limits.

```
# calculating the IQR of the price variable
q1_price = df_clean.price.quantile(.25)
q3_price = df_clean.price.quantile(.75)
IQR_price = q3_price - q1_price
print('IQR of the price variable: ', IQR_price)

# setting the limits
sup_price = q3_price + 1.5 * IQR_price 
inf_price = q1_price - 1.5 * IQR_price

print('Upper price limit: ', sup_price)
print('Lower price limit: ', inf_price)
```

By calculations, the upper bound is 347.5. Therefore, any values ​​above this number are considered superior outliers.

The lower bound is -48.5. As there are no prices below 0, we can conclude that there are no lower outliers in our dataset.

However, care must be taken with null values, given that there would be no logic to renting a property at "zero cost".

The next step is to get the following information from the original dataset:

*   number of entries (rows) of the dataset
*   number of entries (rows) above our upper limit (347.5)
*   the percentage that the outliers represent in the dataset




```
# checking the original dataset
fig, ax = plt.subplots(figsize=(15,3))
df_clean.price.plot(kind='box', vert=False)
ax.set_title('Original Dataset - price')
plt.show()
print("Dataset has {} rows".format(df_clean.shape[0]))
print("{} Entries above 347.5".format(len(df_clean[df_clean.price > 347.5])))
print("Represent {:.2f} % of the dataset".format((len(df_clean[df_clean.price > 347.5]) / df_clean.shape[0]) * 100))
```

As per the analysis above, the dataset has only 4.86% of outliers.

Therefore, I consider that I will not have great losses by deleting this data.


```
# deleting outliers
df_clean.drop(df_clean[df_clean.price > 347.5].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.price == 0].index, axis=0, inplace=True)
```

After deleting, I'll see what the box plot looks like:

```
# price variable box plot
df_clean.price.plot(kind='box', vert=False, figsize=(15,3))
plt.show()
```

By excluding outliers, it is possible to clearly observe the median, the quartiles and the lower and upper limits.

There are still some outliers, but it should be borne in mind that the exclusion of old outliers naturally changed the mean, median, quartiles, and thresholds. See below:


```
# main statistical data of the price variable after deleting outliers
df_clean.price.describe()
```

For a better understanding of the distribution of the data, I will analyze them through a histogram:

**Average of minimum nights to rent (minimum_nights)**

```
# averaging the minimum_nights column
df_clean.minimum_nights.mean()
```

2.78 is the value obtained as an average. However, before adopting it as correct, it is necessary to verify the possible existence of outliers.

Therefore, I will proceed with the analysis of the summary of the statistical data of the *minimum_nights* variable:

```
# summary of the statistical data of the minimum_nights variable
df_clean.minimum_nights.describe()
```

It is possible to notice that the mean (2.78) is much closer to the 3rd quartile (3) than to the median (2). Possibly, there are outliers.

To be sure, I'll visualize the distribution of this data with a box plot:

```
# minimum_nights variable box plot
df_clean.minimum_nights.plot(kind='box', vert=False, figsize=(12,5))
plt.show()
```

From the graph's appearance, it can be seen that there are outliers in the "minimum_nights" variable. So, I chose to clear this data.

For this, I will use the IQR Method:

```
# IQR Method
q1_minimum_nights = df_clean.minimum_nights.quantile(.25)
q3_minimum_nights = df_clean.minimum_nights.quantile(.75)
IQR_minimum_nights = q3_minimum_nights - q1_minimum_nights
print("IQR de minimum_nights: ", IQR_minimum_nights)

# Defining the lower and upper limits
sup_minimum_nights = q3_minimum_nights + 1.5 * IQR_minimum_nights
inf_minimum_nights = q1_minimum_nights - 1.5 * IQR_minimum_nights
print("Upper value of minimum_nights: ", sup_minimum_nights)
print("Lower value of minimum_nights: ", inf_minimum_nights)
```

Now, I'm going to find out how many outliers there are in the minimum_nights variable:

```
# checking the original set
fig, ax = plt.subplots(figsize=(15,3))
df_clean.minimum_nights.plot(kind='box', vert=False)
ax.set_title('Original Dataset - minimum_nights')
plt.show()
print("Dataset has {} rows".format(df_clean.shape[0]))
print("{} entries over 6 nights".format(len(df_clean[df_clean.minimum_nights > 6])))
print("Represent {:.2f} % of the dataset".format((len(df_clean[df_clean.minimum_nights > 6]) / df_clean.minimum_nights.shape[0]) *100))
```

The minimum_nights dataset has 8.29% outliers. I don't think I'll do much harm by deleting this data.

```
# deleting outliers from minimum_nights dataset
df_clean.drop(df_clean[df_clean.minimum_nights > 6].index, axis=0, inplace=True)
```

Once the deletion is done, I will check how the statistical data of the dataset were:

```
# checking the minimum_nights dataset statistical data
df_clean.minimum_nights.describe()
```

Now, I'll check how the box plot turned out:

```
# building the box plot for minimum_nights after cleaning out the outliers
fig, ax = plt.subplots(figsize=(15,3))
df_clean.minimum_nights.plot(kind='box', vert=False)
ax.set_title('Dataset after cleaning - minimum_nights')
plt.show()
print('The Dataset has {} lines'.format(df_clean.shape[0]))
print('{} entries above 6'.format(len(df_clean[df_clean.minimum_nights > 6])))
```

From the *describe()* function, we can conclude that **the average for the minimum number of nights is 1.96 days.**
